## DataFrame Basics

```python
#using python native dictionary we are creating a dataframe like similar structure in dictionary but Dataframe in padnas provides too much additional functionality over the dictionary which we will be building using the lists in below example.

person={
    "first":"Navin",
    "last":"Reddy",
    "email":"navinreddy1222@gmail.com"
}
```

```python
people={
    "first":["Navin"],
    "last":["Reddy"],
    "email":["navinreddy1222@gmail.com"]
}
```

```python
people_new={
    "first":["Navin","Gaurav","Sushil","Harsh"],
    "last":["Reddy","Sharma","Rawal","Rawal"],
    "email":["navinreddy-xyz@gmail.com","gsharma-xyz@gmail.com","sushil-xyz@gmail.com","harsh-xyz@gmail.com"]
}
```

```python
#accessing a key of dictionary
people_new["email"]
```

```python
#creating a dataframe from the dictionary
df=pd.DataFrame(people_new)
```

```python
df
#index is unique values for a row
```

```python
#accessing a column of a dataframe: this is a series object and series has index also.
df["email"]
```

```python
#another way of getting series objects of a dataframe is by using dot
# the good way to get series object or column is by using brackets. because you never know a column name and an inbuilt method name when collides and throws error.
df.email
```

```python
#checking the type of email column a series is basically just a list of data.
# dataframe is a container for multiple of the series objects.
type(df["email"])
```

```python
#accessing a row of a dataframe by index location
df.loc[0]
```

```python
#accessing multiple columns of a dataframe
df[["first","last","email"]]
```

```python
# now that we are fetching multiple columns, this can no longer be a series as series is basically a single column of rows. this will be a dataframe.
type(df[["first","last","email"]])
```

```python
#to get all the columns
df.columns
```

```python
#to get the rows we have to use loc and iloc
#iloc allows us to access location by integer location
```

```python
df.iloc[0]
```

```python
#for getting multiple rows data using iloc
df.iloc[[0,1]]
```

```python
#with iloc we can access columns also. just have to add column index after rows in a list.
df.iloc[[0,1],[1,2]]
```

```python
#loc
df
```

```python
# with iloc we were searching by integer location but with loc we are searching by labels.
df.loc[0]
```

```python
# to get multiple rows
df.loc[[0,1]]
```

```python
# we can also get column using labels not integers as we did previously do it using iloc
df.loc[[0,1,3,2],['last','first']]
```
